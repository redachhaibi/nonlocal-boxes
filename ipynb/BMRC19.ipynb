```python
def chi_X(mu, nu, sigma):
    output=[torch.zeros(2),torch.zeros(2),torch.zeros([2,2])]
    output[0]=torch.tensor([mu])
    output[1]=torch.tensor([nu])
    for a1 in range(2): 
        for a2 in range(2):
            output[2][a1,a2]=(a1+a2+sigma)%2
    return output

def chi_S(mu, nu, sigma, delta, epsilon):
    output=[torch.zeros(2),torch.zeros(2),torch.zeros([2,2])]
    output[mu]=torch.tensor([nu, nu])
    output[(mu+1)%2]=torch.tensor([sigma, (1+sigma)%2])
    for a1 in range(2): 
        for a2 in range(2):
            if mu==0: output[2][a1,a2]=(a2+delta*a1+epsilon)%2
            if mu==1: output[2][a1,a2]=(a1+delta*a2+epsilon)%2
    return output
```
